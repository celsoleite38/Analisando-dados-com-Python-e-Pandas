```python
#criando nosso DAtaframe
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
df.head()
```

```python
#quantidade de linhas e colunas
df.shape
```

```python
#tipos de dados
df.dtypes
```

```python
#qual a receita total
df["Valor Venda"].sum()
```

```python
#qual o custo total?
df["custo"] = df["Custo Unitário"].mul(df["Quantidade"]) #criando a coluna de custo
```

```python
df.head(1)
```

```python
#qual o custo total
round(df["custo"].sum(),2)
```

```python
#agora que temos a receita e custo e o total, podemos achar o lucro total
#vamos criar uma coluna de lucro que sera receita - custo
df["lucro"] = df["Valor Venda"] - df["custo"]
```

```python
df.head(1)
```

```python
#total de lucro
round(df["lucro"].sum(),2)
```

```python
#criando uma colun com total de dias para enviar o produto
df["Tempo_envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
df.head(1)
```

Agora queremos saber a media do tempo de envio para cada MArca, e para isso precisamos trnasformar a coluna TEmpo-envio em númerica

```python
#extraindo apenas os dias
df["Tempo_envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
df.head(1)
```

```python
#verificando o tipo da coluna TEmpo_Envio
df["Tempo_envio"].dtype
```

```python
#media do tempo de envio por MArca
df.groupby("Marca")["Tempo_envio"].mean()
```

Missing Values

```python
#verificando se temos dados faltantes
df.isnull().sum()
```

```python
#vamos agrupar por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["lucro"].sum()
```

```python
#qual o total de produtos vendidos?
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```
